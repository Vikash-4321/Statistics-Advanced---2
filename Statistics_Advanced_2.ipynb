Q1. What is hypothesis testing in statistics?
      
      -Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on a sample of data.
     It helps us check whether an assumption (hypothesis) about a population parameter is likely to be true or not.
     Hypothesis testing is a way to use sample data to test assumptions (claims) about a population, and decide whether the evidence is strong enough to reject the claim.

Q6.: Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10   # number of trials
p = 0.5  # probability of success
size = 1000  # number of samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=np.arange(-0.5, n+1.5, 1), edgecolor='black', alpha=0.7)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

Q7.  : Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


```python
import numpy as np
import scipy.stats as stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
mu_0 = 50   # hypothesized population mean
sigma = 1   # assumed population standard deviation
alpha = 0.05  # significance level

# Sample statistics
n = len(sample_data)
sample_mean = np.mean(sample_data)

# Z-test statistic
z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Results
print(f"Sample Mean = {sample_mean:.3f}")
print(f"Z-statistic = {z_stat:.3f}")
print(f"P-value = {p_value:.4f}")

# Decision
if p_value < alpha:
    print("Reject H0: Sample mean is significantly different from population mean.")
else:
    print("Fail to Reject H0: No significant difference between sample and population mean.")
```

Q8. Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Step 1: Simulate data
np.random.seed(42)  # for reproducibility
mu, sigma, n = 50, 5, 100  # true mean, std deviation, sample size
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate 95% confidence interval
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
alpha = 0.05
df = n - 1

# t critical value for two-tailed test
t_crit = stats.t.ppf(1 - alpha/2, df)

margin_of_error = t_crit * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean = {sample_mean:.3f}")
print(f"95% Confidence Interval = ({ci_lower:.3f}, {ci_upper:.3f})")

# Step 3: Plot histogram of the data
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='--', label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Normal Distribution Sample with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

Q9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.


```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset and plot standardized data.
    """
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # sample standard deviation
    z_scores = (data - mean) / std_dev

    # Print results
    print(f"Mean of original data = {mean:.3f}")
    print(f"Standard deviation of original data = {std_dev:.3f}")
    print(f"First 5 Z-scores: {z_scores[:5]}")

    # Plot histogram of Z-scores
    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7, color='skyblue')
    plt.axvline(0, color='red', linestyle='--', label="Mean (Z=0)")
    plt.title("Histogram of Z-scores (Standardized Data)")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.show()

    return z_scores


# Example usage:
np.random.seed(42)
data = np.random.normal(100, 15, 200)  # Simulate exam scores (mean=100, std=15)
z_scores = calculate_z_scores(data)
```
